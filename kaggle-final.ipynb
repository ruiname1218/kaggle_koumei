```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

print("ðŸš€ PROPER ML MODEL: ã—ã£ã‹ã‚Šã¨ã—ãŸå­¦ç¿’ãƒ¢ãƒ‡ãƒ«")
print("=" * 60)

# ã‚·ãƒ¼ãƒ‰å›ºå®šï¼ˆå†ç¾æ€§ï¼‰
np.random.seed(42)

class LinearRegressionFromScratch:
    """ã‚¹ã‚¯ãƒ©ãƒƒãƒã‹ã‚‰ã®ç·šå½¢å›žå¸°ï¼ˆå­¦ç¿’éŽç¨‹ã‚’è¡¨ç¤ºï¼‰"""
    
    def __init__(self, learning_rate=0.01, max_iterations=1000, tolerance=1e-6):
        self.learning_rate = learning_rate
        self.max_iterations = max_iterations
        self.tolerance = tolerance
        self.weights = None
        self.bias = None
        self.cost_history = []
        
    def _normalize_features(self, X):
        """ç‰¹å¾´é‡ã®æ­£è¦åŒ–"""
        self.feature_means = np.mean(X, axis=0)
        self.feature_stds = np.std(X, axis=0)
        # æ¨™æº–åå·®ãŒ0ã®å ´åˆã¯1ã«ã™ã‚‹ï¼ˆå®šæ•°ç‰¹å¾´é‡å¯¾å¿œï¼‰
        self.feature_stds = np.where(self.feature_stds == 0, 1, self.feature_stds)
        return (X - self.feature_means) / self.feature_stds
    
    def _compute_cost(self, X, y, weights, bias):
        """ã‚³ã‚¹ãƒˆé–¢æ•°ï¼ˆå¹³å‡äºŒä¹—èª¤å·®ï¼‰"""
        m = X.shape[0]
        predictions = X @ weights + bias
        cost = (1/(2*m)) * np.sum((predictions - y)**2)
        return cost
    
    def _compute_gradients(self, X, y, weights, bias):
        """å‹¾é…è¨ˆç®—"""
        m = X.shape[0]
        predictions = X @ weights + bias
        
        dw = (1/m) * X.T @ (predictions - y)
        db = (1/m) * np.sum(predictions - y)
        
        return dw, db
    
    def fit(self, X, y, verbose=True):
        """ãƒ¢ãƒ‡ãƒ«ã®å­¦ç¿’"""
        # ç‰¹å¾´é‡æ­£è¦åŒ–
        X_normalized = self._normalize_features(X)
        
        # ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿åˆæœŸåŒ–
        n_features = X.shape[1]
        self.weights = np.random.normal(0, 0.01, n_features)
        self.bias = 0
        
        self.cost_history = []
        
        if verbose:
            print(f"ðŸ“Š å­¦ç¿’é–‹å§‹: {X.shape[0]}ã‚µãƒ³ãƒ—ãƒ«, {n_features}ç‰¹å¾´é‡")
            print("=" * 50)
        
        # å‹¾é…é™ä¸‹æ³•ã«ã‚ˆã‚‹å­¦ç¿’
        for i in range(self.max_iterations):
            # ã‚³ã‚¹ãƒˆè¨ˆç®—
            cost = self._compute_cost(X_normalized, y, self.weights, self.bias)
            self.cost_history.append(cost)
            
            # å‹¾é…è¨ˆç®—
            dw, db = self._compute_gradients(X_normalized, y, self.weights, self.bias)
            
            # ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿æ›´æ–°
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db
            
            # å­¦ç¿’éŽç¨‹è¡¨ç¤º
            if verbose and (i % 100 == 0 or i == self.max_iterations - 1):
                rmse = np.sqrt(2 * cost)
                print(f"Iteration {i:4d}: Cost = {cost:.6f}, RMSE = {rmse:.4f}")
            
            # åŽæŸåˆ¤å®š
            if i > 0 and abs(self.cost_history[-2] - self.cost_history[-1]) < self.tolerance:
                if verbose:
                    print(f"âœ… åŽæŸã—ã¾ã—ãŸ (Iteration {i})")
                break
        
        return self
    
    def predict(self, X):
        """äºˆæ¸¬"""
        X_normalized = (X - self.feature_means) / self.feature_stds
        return X_normalized @ self.weights + self.bias
    
    def get_rmse(self, X, y):
        """RMSEè¨ˆç®—"""
        predictions = self.predict(X)
        return np.sqrt(np.mean((predictions - y)**2))

def create_enhanced_features(data):
    """æµ·æ´‹åŒ–å­¦ç‰¹å¾´é‡ã‚¨ãƒ³ã‚¸ãƒ‹ã‚¢ãƒªãƒ³ã‚°"""
    enhanced_data = data.copy()
    
    # æ¬ æå€¤ã‚’å¹³å‡ã§è£œå®Œ
    numeric_cols = enhanced_data.select_dtypes(include=[np.number]).columns
    for col in numeric_cols:
        if enhanced_data[col].isnull().any():
            enhanced_data[col] = enhanced_data[col].fillna(enhanced_data[col].mean())
    
    # ðŸ§¬ æµ·æ´‹åŒ–å­¦çš„ã«æ„å‘³ã®ã‚ã‚‹ç‰¹å¾´é‡
    enhanced_data['N_P_ratio'] = enhanced_data['NO3uM'] / (enhanced_data['PO4uM'] + 1e-6)
    enhanced_data['Redfield_deviation'] = np.abs(enhanced_data['N_P_ratio'] - 16.0)
    
    # ðŸŒŠ ç‰©ç†çš„ç›¸äº’ä½œç”¨
    enhanced_data['Depth_Temp_interaction'] = enhanced_data['R_Depth'] * enhanced_data['Temperature_degC'] / 1000
    enhanced_data['Salinity_Oxy_interaction'] = enhanced_data['Salinity1'] * enhanced_data['R_Oxy_micromol.Kg'] / 1000
    
    # ðŸ”¬ æ „é¤Šå¡©é–¢é€£
    enhanced_data['Total_Nitrogen'] = enhanced_data['NO2uM'] + enhanced_data['NO3uM'] + enhanced_data['NH3uM']
    enhanced_data['PO4_normalized'] = enhanced_data['PO4uM'] / (enhanced_data['R_Depth'] + 1) * 100
    
    return enhanced_data

def cross_validate_model(X, y, n_splits=5):
    """K-foldäº¤å·®æ¤œè¨¼"""
    n_samples = len(X)
    fold_size = n_samples // n_splits
    cv_scores = []
    
    print(f"\nðŸ”„ {n_splits}-Fold Cross Validation")
    print("-" * 40)
    
    for fold in range(n_splits):
        # åˆ†å‰²ã‚¤ãƒ³ãƒ‡ãƒƒã‚¯ã‚¹
        val_start = fold * fold_size
        val_end = val_start + fold_size if fold < n_splits - 1 else n_samples
        
        val_idx = list(range(val_start, val_end))
        train_idx = [i for i in range(n_samples) if i not in val_idx]
        
        # ãƒ‡ãƒ¼ã‚¿åˆ†å‰²
        X_train_fold, X_val_fold = X[train_idx], X[val_idx]
        y_train_fold, y_val_fold = y[train_idx], y[val_idx]
        
        # ãƒ¢ãƒ‡ãƒ«å­¦ç¿’
        model = LinearRegressionFromScratch(learning_rate=0.01, max_iterations=1000)
        model.fit(X_train_fold, y_train_fold, verbose=False)
        
        # æ¤œè¨¼ã‚¹ã‚³ã‚¢
        val_score = model.get_rmse(X_val_fold, y_val_fold)
        cv_scores.append(val_score)
        
        print(f"Fold {fold + 1}: RMSE = {val_score:.4f}")
    
    mean_score = np.mean(cv_scores)
    std_score = np.std(cv_scores)
    print(f"\nðŸ“Š CVçµæžœ: {mean_score:.4f} Â± {std_score:.4f}")
    
    return cv_scores

# ðŸš€ ãƒ¡ã‚¤ãƒ³å®Ÿè¡Œ
print("ðŸ“Š ãƒ‡ãƒ¼ã‚¿å‰å‡¦ç†...")

# ãƒ‡ãƒ¼ã‚¿ã‚¯ãƒªãƒ¼ãƒ‹ãƒ³ã‚°
cols_to_drop = [c for c in ["Unnamed: 12", "id"] if c in train.columns]
if cols_to_drop:
    train_clean = train.drop(columns=cols_to_drop)
else:
    train_clean = train.copy()

# ãƒ†ã‚¹ãƒˆIDã‚’ä¿å­˜
if "id" in test.columns:
    test_id = test["id"].copy()
    test_clean = test.drop(columns=["id"])
else:
    test_clean = test.copy()
    test_id = range(1455, 1455 + len(test))

# æ‹¡å¼µç‰¹å¾´é‡ä½œæˆ
print("ðŸ§¬ ç‰¹å¾´é‡ã‚¨ãƒ³ã‚¸ãƒ‹ã‚¢ãƒªãƒ³ã‚°...")
train_enhanced = create_enhanced_features(train_clean)
test_enhanced = create_enhanced_features(test_clean)

# æœ€é©ç‰¹å¾´é‡é¸æŠž
selected_features = [
    'PO4uM', 'NO3uM', 'R_Oxy_micromol.Kg', 'Salinity1', 'Temperature_degC',
    'N_P_ratio', 'Depth_Temp_interaction', 'Salinity_Oxy_interaction', 
    'Total_Nitrogen', 'PO4_normalized'
]

print(f"ðŸŽ¯ é¸æŠžç‰¹å¾´é‡: {len(selected_features)}å€‹")
for i, feature in enumerate(selected_features):
    print(f"  {i+1:2d}. {feature}")

# ãƒ‡ãƒ¼ã‚¿æº–å‚™
X = train_enhanced[selected_features].values
y = train_enhanced['DIC'].values
X_test = test_enhanced[selected_features].values

print(f"\nðŸ“ˆ ãƒ‡ãƒ¼ã‚¿å½¢çŠ¶:")
print(f"  å­¦ç¿’ãƒ‡ãƒ¼ã‚¿: {X.shape}")
print(f"  ãƒ†ã‚¹ãƒˆãƒ‡ãƒ¼ã‚¿: {X_test.shape}")
print(f"  ã‚¿ãƒ¼ã‚²ãƒƒãƒˆç¯„å›²: {y.min():.1f} - {y.max():.1f}")

# Cross Validation
cv_scores = cross_validate_model(X, y, n_splits=5)

# æœ€çµ‚ãƒ¢ãƒ‡ãƒ«å­¦ç¿’ï¼ˆå…¨ãƒ‡ãƒ¼ã‚¿ï¼‰
print(f"\nðŸš€ æœ€çµ‚ãƒ¢ãƒ‡ãƒ«å­¦ç¿’ï¼ˆå…¨ãƒ‡ãƒ¼ã‚¿ï¼‰")
print("=" * 50)

final_model = LinearRegressionFromScratch(learning_rate=0.01, max_iterations=1500)
final_model.fit(X, y, verbose=True)

# æœ€çµ‚è©•ä¾¡
final_rmse = final_model.get_rmse(X, y)
print(f"\nðŸ“Š æœ€çµ‚å­¦ç¿’çµæžœ:")
print(f"  å­¦ç¿’ãƒ‡ãƒ¼ã‚¿RMSE: {final_rmse:.4f}")
print(f"  CVå¹³å‡RMSE: {np.mean(cv_scores):.4f}")

# å­¦ç¿’æ›²ç·šã®ç°¡æ˜“è¡¨ç¤º
cost_history = final_model.cost_history
print(f"\nðŸ“ˆ å­¦ç¿’éŽç¨‹:")
print(f"  åˆæœŸCost: {cost_history[0]:.6f}")
print(f"  æœ€çµ‚Cost: {cost_history[-1]:.6f}")
print(f"  æ”¹å–„: {((cost_history[0] - cost_history[-1])/cost_history[0]*100):.1f}%")

# ãƒ†ã‚¹ãƒˆäºˆæ¸¬
test_predictions = final_model.predict(X_test)

print(f"\nðŸ”® ãƒ†ã‚¹ãƒˆäºˆæ¸¬:")
print(f"  äºˆæ¸¬ç¯„å›²: {test_predictions.min():.1f} - {test_predictions.max():.1f}")
print(f"  äºˆæ¸¬å¹³å‡: {test_predictions.mean():.1f}")

# ãƒ¢ãƒ‡ãƒ«é‡è¦åº¦ï¼ˆçµ¶å¯¾å€¤ï¼‰
feature_importance = np.abs(final_model.weights)
importance_ranking = np.argsort(feature_importance)[::-1]

print(f"\nðŸ” ç‰¹å¾´é‡é‡è¦åº¦ï¼ˆä¸Šä½5ä½ï¼‰:")
for i in range(min(5, len(selected_features))):
    idx = importance_ranking[i]
    print(f"  {i+1}. {selected_features[idx]:25s}: {feature_importance[idx]:.4f}")

# æå‡ºãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ ä½œæˆ
submission = pd.DataFrame({
    "id": test_id,
    "DIC": test_predictions
})

print(f"\nâœ… å­¦ç¿’å®Œäº†ï¼äºˆæƒ³ã‚¹ã‚³ã‚¢: ~{np.mean(cv_scores):.1f}")
```

```python
# ã‚¢ãƒ³ã‚µãƒ³ãƒ–ãƒ«äºˆæ¸¬ï¼ˆK-foldã®å…¨ãƒ¢ãƒ‡ãƒ«ã®å¹³å‡ï¼‰
def ensemble_predict(models, X_test):
    """K-foldã®å…¨ãƒ¢ãƒ‡ãƒ«ã§ã‚¢ãƒ³ã‚µãƒ³ãƒ–ãƒ«äºˆæ¸¬"""
    predictions = []
    
    for fold, (model, scaler, imputer) in enumerate(models):
        # ãƒ†ã‚¹ãƒˆãƒ‡ãƒ¼ã‚¿ã®å‰å‡¦ç†
        X_test_processed = imputer.transform(X_test)
        X_test_processed = scaler.transform(X_test_processed)
        X_test_tensor = torch.tensor(X_test_processed, dtype=torch.float32)
        
        # äºˆæ¸¬
        model.eval()
        with torch.no_grad():
            fold_pred = model(X_test_tensor).squeeze().numpy()
        
        predictions.append(fold_pred)
        print(f"Fold {fold + 1} prediction range: {fold_pred.min():.2f} - {fold_pred.max():.2f}")
    
    # ã‚¢ãƒ³ã‚µãƒ³ãƒ–ãƒ«ï¼ˆå¹³å‡ï¼‰
    ensemble_pred = np.mean(predictions, axis=0)
    return ensemble_pred, predictions

# ã‚¢ãƒ³ã‚µãƒ³ãƒ–ãƒ«äºˆæ¸¬ã®å®Ÿè¡Œ
ensemble_predictions, individual_predictions = ensemble_predict(models, X_test)

print(f"\nã‚¢ãƒ³ã‚µãƒ³ãƒ–ãƒ«äºˆæ¸¬ç¯„å›²: {ensemble_predictions.min():.2f} - {ensemble_predictions.max():.2f}")

# æå‡ºãƒ•ã‚¡ã‚¤ãƒ«ã®ä½œæˆ
submission = pd.DataFrame({
    "id": test_id if 'test_id' in locals() else range(1455, 1455 + len(ensemble_predictions)), 
    "DIC": ensemble_predictions
})

submission.to_csv("submission_kfold.csv", index=False)
print("\næå‡ºãƒ•ã‚¡ã‚¤ãƒ« 'submission_kfold.csv' ã‚’ä½œæˆã—ã¾ã—ãŸã€‚")
print(f"æå‡ºãƒ‡ãƒ¼ã‚¿çµ±è¨ˆ:")
print(f"  å¹³å‡: {ensemble_predictions.mean():.2f}")
print(f"  æ¨™æº–åå·®: {ensemble_predictions.std():.2f}")
print(f"  æœ€å°å€¤: {ensemble_predictions.min():.2f}")
print(f"  æœ€å¤§å€¤: {ensemble_predictions.max():.2f}")

# äºˆæ¸¬ã®åˆ†æ•£ï¼ˆä¸ç¢ºå®Ÿæ€§ï¼‰ã‚‚è¨ˆç®—
prediction_std = np.std(individual_predictions, axis=0)
print(f"\näºˆæ¸¬ã®ä¸ç¢ºå®Ÿæ€§ï¼ˆæ¨™æº–åå·®ï¼‰:")
print(f"  å¹³å‡: {prediction_std.mean():.2f}")
print(f"  æœ€å¤§: {prediction_std.max():.2f}")

submission.head()
```
