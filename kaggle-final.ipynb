```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from sklearn.model_selection import KFold, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import random

# シード値固定
def fix_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

fix_seed(42)

# 1) 列の整理
cols_to_drop = [c for c in ["Unnamed: 12", "id"] if c in train.columns]
if cols_to_drop:
    train = train.drop(columns=cols_to_drop)

# test には 'id' があるなら、学習時には不要なので落としてもOK（提出で使うなら残す）
if "id" in test.columns:
    test_id = test["id"].copy()   # 提出に使うなら退避
    test = test.drop(columns=["id"])

# 2) 共通列を揃える（順序も train に合わせる）
feature_cols = train.drop(columns=["DIC"]).columns
common_cols = [c for c in feature_cols if c in test.columns]
X = train[common_cols].copy()
y = train["DIC"].copy()
X_test = test[common_cols].copy()

# 3) 特徴量エンジニアリング
# 相関の高い特徴量から新しい特徴量を作成
X['Temp_Sal_ratio'] = X['Temperature_degC'] / (X['Salinity1'] + 1e-8)
X['Depth_Temp_interaction'] = X['R_Depth'] * X['Temperature_degC']
X['Nutrient_sum'] = X['NO2uM'] + X['NO3uM'] + X['NH3uM'] + X['PO4uM']
X['Oxygen_Depth_ratio'] = X['R_Oxy_micromol.Kg'] / (X['R_Depth'] + 1e-8)

# テストデータにも同様の特徴量を追加
X_test['Temp_Sal_ratio'] = X_test['Temperature_degC'] / (X_test['Salinity1'] + 1e-8)
X_test['Depth_Temp_interaction'] = X_test['R_Depth'] * X_test['Temperature_degC']
X_test['Nutrient_sum'] = X_test['NO2uM'] + X_test['NO3uM'] + X_test['NH3uM'] + X_test['PO4uM']
X_test['Oxygen_Depth_ratio'] = X_test['R_Oxy_micromol.Kg'] / (X_test['R_Depth'] + 1e-8)

# 4) 異常値処理（外れ値をクリッピング）
def clip_outliers(data, columns, lower_percentile=1, upper_percentile=99):
    data_clipped = data.copy()
    for col in columns:
        lower_bound = np.percentile(data[col], lower_percentile)
        upper_bound = np.percentile(data[col], upper_percentile)
        data_clipped[col] = np.clip(data[col], lower_bound, upper_bound)
    return data_clipped

numeric_cols = X.select_dtypes(include=[np.number]).columns
X = clip_outliers(X, numeric_cols)

print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"Test shape: {X_test.shape}")
```

```python
import torch
import torch.nn as nn

class MLPModel(nn.Module):
    def __init__(self, input_size, dropout_rate=0.3):
        super(MLPModel, self).__init__()
        
        # より深いネットワーク + Dropout + BatchNorm
        self.fc1 = nn.Linear(input_size, 512)
        self.bn1 = nn.BatchNorm1d(512)
        self.act1 = nn.ReLU()
        self.dropout1 = nn.Dropout(dropout_rate)
        
        self.fc2 = nn.Linear(512, 256)
        self.bn2 = nn.BatchNorm1d(256)
        self.act2 = nn.ReLU()
        self.dropout2 = nn.Dropout(dropout_rate)
        
        self.fc3 = nn.Linear(256, 128)
        self.bn3 = nn.BatchNorm1d(128)
        self.act3 = nn.ReLU()
        self.dropout3 = nn.Dropout(dropout_rate)
        
        # 出力層
        self.fc4 = nn.Linear(128, 1)
        
        # 重み初期化
        self._initialize_weights()
        
    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_normal_(m.weight)
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
                    
    def forward(self, x):
        # 順伝播（変更なし）
        x = self.fc1(x)
        x = self.bn1(x)
        x = self.act1(x)
        x = self.dropout1(x)
        
        x = self.fc2(x)
        x = self.bn2(x)
        x = self.act2(x)
        x = self.dropout2(x)
        
        x = self.fc3(x)
        x = self.bn3(x)
        x = self.act3(x)
        x = self.dropout3(x)
        
        x = self.fc4(x)
        return x
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')

class OceanChemistryDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

def train_fold(X_train, y_train, X_val, y_val, input_size, epochs=2000):
    """1つのfoldの学習"""
    fix_seed(42)  # 各foldでシード値固定
    
    # データセットとローダー作成
    train_dataset = OceanChemistryDataset(X_train, y_train)
    val_dataset = OceanChemistryDataset(X_val, y_val)
    
    train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)
    
    # モデル初期化
    model = MLPModel(input_size=input_size)
    
    # 損失関数とオプティマイザー
    criterion = nn.SmoothL1Loss()
    optimizer = optim.AdamW(model.parameters(), lr=0.001, weight_decay=1e-5)
    
    # 学習率スケジューラー
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode="min", factor=0.7, patience=100, verbose=False
    )
    
    best_val_loss = float('inf')
    patience = 200
    patience_counter = 0
    
    for epoch in range(epochs):
        # 学習フェーズ
        model.train()
        running_loss = 0.0
        
        for X_batch, y_batch in train_loader:
            optimizer.zero_grad()
            outputs = model(X_batch)
            loss = criterion(outputs.squeeze(), y_batch)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        
        # 検証フェーズ
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for X_batch, y_batch in val_loader:
                outputs = model(X_batch)
                loss = criterion(outputs.squeeze(), y_batch)
                val_loss += loss.item()
        
        train_loss_epoch = running_loss / len(train_loader)
        val_loss_epoch = val_loss / len(val_loader)
        
        scheduler.step(val_loss_epoch)
        
        # Early stopping
        if val_loss_epoch < best_val_loss:
            best_val_loss = val_loss_epoch
            patience_counter = 0
            best_model_state = model.state_dict().copy()
        else:
            patience_counter += 1
            
        if patience_counter >= patience:
            break
            
        if epoch % 200 == 0:
            print(f"  Epoch {epoch:4d}: Train Loss: {train_loss_epoch:.6f}, Val Loss: {val_loss_epoch:.6f}")
    
    # ベストモデルをロード
    model.load_state_dict(best_model_state)
    return model, best_val_loss

# K-fold交差検証の実行
n_folds = 5
kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)

models = []
val_scores = []

print("K-fold Cross Validation開始...")
for fold, (train_idx, val_idx) in enumerate(kf.split(X)):
    print(f"\n=== Fold {fold + 1}/{n_folds} ===")
    
    X_train_fold = X.iloc[train_idx]
    y_train_fold = y.iloc[train_idx]
    X_val_fold = X.iloc[val_idx]
    y_val_fold = y.iloc[val_idx]
    
    # 前処理（各foldで独立して実行）
    imputer = SimpleImputer(strategy="mean")
    X_train_fold = imputer.fit_transform(X_train_fold)
    X_val_fold = imputer.transform(X_val_fold)
    
    scaler = StandardScaler()
    X_train_fold = scaler.fit_transform(X_train_fold)
    X_val_fold = scaler.transform(X_val_fold)
    
    # モデル学習
    model, val_loss = train_fold(
        X_train_fold, y_train_fold.values, 
        X_val_fold, y_val_fold.values, 
        input_size=X_train_fold.shape[1]
    )
    
    models.append((model, scaler, imputer))
    val_scores.append(val_loss)
    
    print(f"Fold {fold + 1} validation score: {val_loss:.6f}")

print(f"\n=== K-fold結果 ===")
print(f"平均validation score: {np.mean(val_scores):.6f} ± {np.std(val_scores):.6f}")
for i, score in enumerate(val_scores):
    print(f"Fold {i+1}: {score:.6f}")
```

```python
# アンサンブル予測（K-foldの全モデルの平均）
def ensemble_predict(models, X_test):
    """K-foldの全モデルでアンサンブル予測"""
    predictions = []
    
    for fold, (model, scaler, imputer) in enumerate(models):
        # テストデータの前処理
        X_test_processed = imputer.transform(X_test)
        X_test_processed = scaler.transform(X_test_processed)
        X_test_tensor = torch.tensor(X_test_processed, dtype=torch.float32)
        
        # 予測
        model.eval()
        with torch.no_grad():
            fold_pred = model(X_test_tensor).squeeze().numpy()
        
        predictions.append(fold_pred)
        print(f"Fold {fold + 1} prediction range: {fold_pred.min():.2f} - {fold_pred.max():.2f}")
    
    # アンサンブル（平均）
    ensemble_pred = np.mean(predictions, axis=0)
    return ensemble_pred, predictions

# アンサンブル予測の実行
ensemble_predictions, individual_predictions = ensemble_predict(models, X_test)

print(f"\nアンサンブル予測範囲: {ensemble_predictions.min():.2f} - {ensemble_predictions.max():.2f}")

# 提出ファイルの作成
submission = pd.DataFrame({
    "id": test_id if 'test_id' in locals() else range(1455, 1455 + len(ensemble_predictions)), 
    "DIC": ensemble_predictions
})

submission.to_csv("submission_kfold.csv", index=False)
print("\n提出ファイル 'submission_kfold.csv' を作成しました。")
print(f"提出データ統計:")
print(f"  平均: {ensemble_predictions.mean():.2f}")
print(f"  標準偏差: {ensemble_predictions.std():.2f}")
print(f"  最小値: {ensemble_predictions.min():.2f}")
print(f"  最大値: {ensemble_predictions.max():.2f}")

# 予測の分散（不確実性）も計算
prediction_std = np.std(individual_predictions, axis=0)
print(f"\n予測の不確実性（標準偏差）:")
print(f"  平均: {prediction_std.mean():.2f}")
print(f"  最大: {prediction_std.max():.2f}")

submission.head()
```
