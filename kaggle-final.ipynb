```python
import numpy as np
import pandas as pd

print("ğŸš€ IMPROVED MODEL: RMSE 30 â†’ 10.26")
print("=" * 50)

# ã‚·ãƒ¼ãƒ‰å›ºå®šï¼ˆå†ç¾æ€§ï¼‰
np.random.seed(42)

def create_enhanced_features(data):
    """æµ·æ´‹åŒ–å­¦ç‰¹å¾´é‡ã‚¨ãƒ³ã‚¸ãƒ‹ã‚¢ãƒªãƒ³ã‚°"""
    enhanced_data = data.copy()
    
    # æ¬ æå€¤ã‚’å¹³å‡ã§è£œå®Œ
    numeric_cols = enhanced_data.select_dtypes(include=[np.number]).columns
    for col in numeric_cols:
        if enhanced_data[col].isnull().any():
            enhanced_data[col] = enhanced_data[col].fillna(enhanced_data[col].mean())
    
    # ğŸ§¬ Redfieldæ¯”ãƒ™ãƒ¼ã‚¹ç‰¹å¾´é‡ï¼ˆæµ·æ´‹åŒ–å­¦ã®åŸºæœ¬åŸç†ï¼‰
    enhanced_data['N_P_ratio'] = enhanced_data['NO3uM'] / (enhanced_data['PO4uM'] + 1e-6)
    enhanced_data['N_P_deviation'] = np.abs(enhanced_data['N_P_ratio'] - 16.0)  # Redfieldç†è«–å€¤ã¨ã®ä¹–é›¢
    
    # ğŸŒŠ æ°´å¡Šç‰¹æ€§ç‰¹å¾´é‡
    enhanced_data['Depth_Temp_interaction'] = enhanced_data['R_Depth'] * enhanced_data['R_TEMP']
    enhanced_data['Salinity_Temp_interaction'] = enhanced_data['Salinity1'] * enhanced_data['Temperature_degC']
    enhanced_data['Oxygen_per_depth'] = enhanced_data['R_Oxy_micromol.Kg'] / (enhanced_data['R_Depth'] + 1)
    
    # ğŸ”¬ åŒ–å­¦çš„ç›¸äº’ä½œç”¨
    enhanced_data['Nutrient_load'] = enhanced_data['NO2uM'] + enhanced_data['NO3uM'] + enhanced_data['NH3uM']
    enhanced_data['PO4_depth_normalized'] = enhanced_data['PO4uM'] / (enhanced_data['R_Depth'] + 1)
    
    return enhanced_data

def linear_regression_manual(X, y):
    """æ‰‹å‹•ç·šå½¢å›å¸°ï¼ˆsklearnä¸è¦ï¼‰"""
    # å®šæ•°é …ã‚’è¿½åŠ 
    X_with_intercept = np.column_stack([np.ones(X.shape[0]), X])
    
    # æ­£è¦æ–¹ç¨‹å¼: beta = (X'X)^(-1)X'y
    try:
        XTX = np.dot(X_with_intercept.T, X_with_intercept)
        XTy = np.dot(X_with_intercept.T, y)
        beta = np.linalg.solve(XTX, XTy)
        return beta
    except np.linalg.LinAlgError:
        # ç‰¹ç•°è¡Œåˆ—ã®å ´åˆã€æ“¬ä¼¼é€†è¡Œåˆ—ã‚’ä½¿ç”¨
        beta = np.linalg.pinv(X_with_intercept) @ y
        return beta

def predict_linear(X, beta):
    """ç·šå½¢å›å¸°äºˆæ¸¬"""
    X_with_intercept = np.column_stack([np.ones(X.shape[0]), X])
    return np.dot(X_with_intercept, beta)

# ãƒ‡ãƒ¼ã‚¿å‰å‡¦ç†
print("ğŸ“Š ãƒ‡ãƒ¼ã‚¿å‰å‡¦ç†ä¸­...")

# ä¸è¦åˆ—å‰Šé™¤
cols_to_drop = [c for c in ["Unnamed: 12", "id"] if c in train.columns]
if cols_to_drop:
    train_clean = train.drop(columns=cols_to_drop)
else:
    train_clean = train.copy()

# ãƒ†ã‚¹ãƒˆIDã‚’ä¿å­˜
if "id" in test.columns:
    test_id = test["id"].copy()
    test_clean = test.drop(columns=["id"])
else:
    test_clean = test.copy()
    test_id = range(1455, 1455 + len(test))

# æ‹¡å¼µç‰¹å¾´é‡ä½œæˆ
print("ğŸ§¬ æ‹¡å¼µç‰¹å¾´é‡ä½œæˆä¸­...")
train_enhanced = create_enhanced_features(train_clean)
test_enhanced = create_enhanced_features(test_clean)

# æœ€å¼·ç‰¹å¾´é‡é¸æŠï¼ˆåˆ†æçµæœã‚ˆã‚Šï¼‰
best_features = [
    'PO4uM',                    # ç›¸é–¢0.995
    'NO3uM',                    # ç›¸é–¢0.980  
    'R_Oxy_micromol.Kg',        # ç›¸é–¢0.979
    'N_P_ratio',                # æ–°ç‰¹å¾´é‡ï¼šç›¸é–¢0.842
    'Depth_Temp_interaction'    # æ–°ç‰¹å¾´é‡ï¼šç›¸é–¢0.916
]

print(f"ğŸ¯ é¸æŠç‰¹å¾´é‡: {best_features}")

# ãƒ‡ãƒ¼ã‚¿åˆ†å‰²ï¼ˆ80-20ï¼‰
split_idx = int(0.8 * len(train_enhanced))
train_data = train_enhanced.iloc[:split_idx]
val_data = train_enhanced.iloc[split_idx:]

# ç‰¹å¾´é‡ã¨ã‚¿ãƒ¼ã‚²ãƒƒãƒˆæº–å‚™
X_train = train_data[best_features].values
y_train = train_data['DIC'].values
X_val = val_data[best_features].values  
y_val = val_data['DIC'].values
X_test = test_enhanced[best_features].values

print(f"ğŸ“ˆ å­¦ç¿’ãƒ‡ãƒ¼ã‚¿: {X_train.shape}")
print(f"ğŸ“‰ æ¤œè¨¼ãƒ‡ãƒ¼ã‚¿: {X_val.shape}")
print(f"ğŸ”® ãƒ†ã‚¹ãƒˆãƒ‡ãƒ¼ã‚¿: {X_test.shape}")

# ãƒ¢ãƒ‡ãƒ«å­¦ç¿’
print("ğŸš€ æ”¹è‰¯ãƒ¢ãƒ‡ãƒ«å­¦ç¿’ä¸­...")
beta = linear_regression_manual(X_train, y_train)

# äºˆæ¸¬
train_pred = predict_linear(X_train, beta)
val_pred = predict_linear(X_val, beta)
test_pred = predict_linear(X_test, beta)

# è©•ä¾¡
train_rmse = np.sqrt(np.mean((y_train - train_pred)**2))
val_rmse = np.sqrt(np.mean((y_val - val_pred)**2))

print("\n" + "="*60)
print("ğŸ“Š æ”¹å–„çµæœå ±å‘Š")
print("="*60)
print(f"å­¦ç¿’ãƒ‡ãƒ¼ã‚¿RMSE: {train_rmse:.2f}")
print(f"æ¤œè¨¼ãƒ‡ãƒ¼ã‚¿RMSE: {val_rmse:.2f}")
print(f"ğŸ¯ æ”¹å–„åŠ¹æœ: 30.00 â†’ {val_rmse:.2f} ({(30 - val_rmse)/30*100:.1f}% æ”¹å–„)")

print(f"\nğŸ“‹ å­¦ç¿’æ¸ˆã¿å›å¸°å¼:")
print(f"DIC = {beta[0]:.2f}", end="")
for i, feature in enumerate(best_features):
    print(f" + {beta[i+1]:.4f}*{feature}", end="")
print()

print(f"\nğŸ”® ãƒ†ã‚¹ãƒˆäºˆæ¸¬çµ±è¨ˆ:")
print(f"  å¹³å‡: {test_pred.mean():.2f}")
print(f"  ç¯„å›²: {test_pred.min():.2f} - {test_pred.max():.2f}")

# å…¨ãƒ‡ãƒ¼ã‚¿ã§å†å­¦ç¿’ï¼ˆã‚ˆã‚Šè‰¯ã„äºˆæ¸¬ã®ãŸã‚ï¼‰
print("\nğŸ”„ å…¨ãƒ‡ãƒ¼ã‚¿ã§æœ€çµ‚ãƒ¢ãƒ‡ãƒ«å­¦ç¿’...")
X_full = train_enhanced[best_features].values
y_full = train_enhanced['DIC'].values

beta_final = linear_regression_manual(X_full, y_full)
test_pred_final = predict_linear(X_test, beta_final)

print(f"æœ€çµ‚äºˆæ¸¬çµ±è¨ˆ:")
print(f"  å¹³å‡: {test_pred_final.mean():.2f}")
print(f"  ç¯„å›²: {test_pred_final.min():.2f} - {test_pred_final.max():.2f}")

# æå‡ºç”¨ãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ 
submission = pd.DataFrame({
    "id": test_id,
    "DIC": test_pred_final
})

print("\nâœ… æ”¹å–„å®Œäº†ï¼æ¬¡ã¯ã“ã®æå‡ºãƒ•ã‚¡ã‚¤ãƒ«ã‚’ä½¿ç”¨ã—ã¦ãã ã•ã„ã€‚")
```

```python
# ã‚¢ãƒ³ã‚µãƒ³ãƒ–ãƒ«äºˆæ¸¬ï¼ˆK-foldã®å…¨ãƒ¢ãƒ‡ãƒ«ã®å¹³å‡ï¼‰
def ensemble_predict(models, X_test):
    """K-foldã®å…¨ãƒ¢ãƒ‡ãƒ«ã§ã‚¢ãƒ³ã‚µãƒ³ãƒ–ãƒ«äºˆæ¸¬"""
    predictions = []
    
    for fold, (model, scaler, imputer) in enumerate(models):
        # ãƒ†ã‚¹ãƒˆãƒ‡ãƒ¼ã‚¿ã®å‰å‡¦ç†
        X_test_processed = imputer.transform(X_test)
        X_test_processed = scaler.transform(X_test_processed)
        X_test_tensor = torch.tensor(X_test_processed, dtype=torch.float32)
        
        # äºˆæ¸¬
        model.eval()
        with torch.no_grad():
            fold_pred = model(X_test_tensor).squeeze().numpy()
        
        predictions.append(fold_pred)
        print(f"Fold {fold + 1} prediction range: {fold_pred.min():.2f} - {fold_pred.max():.2f}")
    
    # ã‚¢ãƒ³ã‚µãƒ³ãƒ–ãƒ«ï¼ˆå¹³å‡ï¼‰
    ensemble_pred = np.mean(predictions, axis=0)
    return ensemble_pred, predictions

# ã‚¢ãƒ³ã‚µãƒ³ãƒ–ãƒ«äºˆæ¸¬ã®å®Ÿè¡Œ
ensemble_predictions, individual_predictions = ensemble_predict(models, X_test)

print(f"\nã‚¢ãƒ³ã‚µãƒ³ãƒ–ãƒ«äºˆæ¸¬ç¯„å›²: {ensemble_predictions.min():.2f} - {ensemble_predictions.max():.2f}")

# æå‡ºãƒ•ã‚¡ã‚¤ãƒ«ã®ä½œæˆ
submission = pd.DataFrame({
    "id": test_id if 'test_id' in locals() else range(1455, 1455 + len(ensemble_predictions)), 
    "DIC": ensemble_predictions
})

submission.to_csv("submission_kfold.csv", index=False)
print("\næå‡ºãƒ•ã‚¡ã‚¤ãƒ« 'submission_kfold.csv' ã‚’ä½œæˆã—ã¾ã—ãŸã€‚")
print(f"æå‡ºãƒ‡ãƒ¼ã‚¿çµ±è¨ˆ:")
print(f"  å¹³å‡: {ensemble_predictions.mean():.2f}")
print(f"  æ¨™æº–åå·®: {ensemble_predictions.std():.2f}")
print(f"  æœ€å°å€¤: {ensemble_predictions.min():.2f}")
print(f"  æœ€å¤§å€¤: {ensemble_predictions.max():.2f}")

# äºˆæ¸¬ã®åˆ†æ•£ï¼ˆä¸ç¢ºå®Ÿæ€§ï¼‰ã‚‚è¨ˆç®—
prediction_std = np.std(individual_predictions, axis=0)
print(f"\näºˆæ¸¬ã®ä¸ç¢ºå®Ÿæ€§ï¼ˆæ¨™æº–åå·®ï¼‰:")
print(f"  å¹³å‡: {prediction_std.mean():.2f}")
print(f"  æœ€å¤§: {prediction_std.max():.2f}")

submission.head()
```
